```python
T = int(input())

for i in range(T):
    s = input()
    stack = []
    is_vps = True

    for ch in s:
        if ch == '(':
            stack.append(ch)
        else:
            if len(stack) == 0:
                is_vps = False
                break
            stack.pop()

    if len(stack) != 0:
        is_vps = False

    if is_vps:
        print('YES')
    else:
        print('NO')
```
