```python
N =  int(input())
import random
def make_matrix(N):
    matrix = []

    for i in range(N):
        row=[]
        for j in range(N):
            n = random.randint(1, N*N*10-1)
            row.append(n)
        matrix.append(row)

    return matrix

def print_matrix(matrix):
    n = len(matrix)

    for i in range(n):
        for j in range(n):
            print(f"{matrix[i][j]:3}", end="")
            if j == n-1:
                print("")

A = make_matrix(N)
B = make_matrix(N)
C = make_matrix(N)

matrix = []

for i in range(N):
    row = []
    for j in range(N):
        summary = 0
        for k in range(N):
            summary += A[i][k]*B[k][j]    
        row.append(summary)
    matrix.append(row)

AB = matrix

matrix = []

for i in range(N):
    row = []
    for j in range(N):
        summary = 0
        summary += AB[i][j]+C[i][j]
        row.append(summary)
    matrix.append(row)

print_matrix(matrix)
```
