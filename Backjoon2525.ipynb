```python
hour, minute = map(int, input().split())
addedminute = int(input())
hour = hour*60
totalminute = hour + minute + addedminute
hour = totalminute//60
minute = totalminute%60

if hour > 23:
  hour - 24
else hour =< 23
  hour = hour

print(hour, minute)
```
