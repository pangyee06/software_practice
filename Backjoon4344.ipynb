```python
T = int(input())

for i in range(T):
    data = list(map(int, input().split()))
    n = data[0]
    scores = data[1:]

    avg = sum(scores) / n
    count = 0

    for score in scores:
        if score > avg:
            count += 1

    result = count / n * 100
    print(f"{result:.3f}%")
```
