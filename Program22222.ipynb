```python
N =  int(input())
import random
def make_matrix(N):
    matrix = []

    for i in range(N):
        row=[]
        for j in range(N):
            n = random.randint(1, N*N*10-1)
            row.append(n)
        matrix.append(row)

    return matrix

def print_matrix(matrix):
    n = len(matrix)

    for i in range(n):
        for j in range(n):
            print(f"{matrix[i][j]:3}", end="")
            if j == n-1:
                print("")

original_matrix = make_matrix(N)
matrix = []

for i in range(N):
    row = []
    for j in range(N):
        row.append(original_matrix[j][i])
    matrix.append(row)

print_matrix(matrix)
        
```
